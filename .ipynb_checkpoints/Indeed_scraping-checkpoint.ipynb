```python
import csv
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.keys import Keys
#La classe By est utilisée pour localiser des éléments dans un document.
from  selenium.webdriver.common.by import By
from time import sleep
from selenium.common.exceptions import NoSuchElementException
```

```python
driver=webdriver.Chrome()
def get_url(terme_search):
    template="https://ma.indeed.com/jobs?q={}"
    terme_search=terme_search.replace(' ','+')
    #ville=ville.replace(' ','+')
    
    url=template.format(terme_search)
    url += "&start={}"
    return url
tab=[]
url=get_url('data engineer')
tbd=[]
for page in range(0,25):
    driver.get(url.format(page*10))
    sleep(6)
    jobs=driver.find_elements(By.XPATH,'//div[@class="job_seen_beacon"]')
    for item in jobs:
        item.click()
        sleep(6)
        job={}  
        try:
            job["title"]=driver.find_element(By.XPATH,'//h2[@class="icl-u-xs-mb--xs icl-u-xs-mt--none jobsearch-JobInfoHeader-title is-embedded"]').text
        except AttributeError:
            job["title"]=None
        try:
            job["company"]=driver.find_element(By.XPATH,"//div[@data-company-name='true']").text
        except AttributeError:
            job["company"]=None
        try:
            job["location"]=driver.find_element(By.XPATH,'//div[@class="icl-u-xs-mt--xs is-embedded jobsearch-JobInfoHeader-subtitle icl-u-xs-mb--md"]/div[2]').text
        except AttributeError:
            job["location"]=None   
        try:
            job["type_contrat"]=driver.find_element(By.XPATH,'//div[@id="jobDetailsSection"]/div/div[2]').text 
        except NoSuchElementException:
            job["type_contrat"]=None
        try:
            job["date_publication"]=driver.find_element(By.XPATH,'//li[@class="css-5vsc1i eu4oa1w0"]/span[2]').text 
        except AttributeError:
            job["date_publication"]=None
        try:
            job["description"]=driver.find_element(By.XPATH,'//div[@id="jobDescriptionText"]').text
        except AttributeError:
            job["description"]=None
        tab.append(job)
```

```python
data['date_publication'] = data['date_publication'].str.replace(r'[^\w\s]', '')
data["date_publication"]
```

```python
import pandas as pd
import datetime



# Boucle sur les lignes du DataFrame
for i, row in data.iterrows():
    publication_date_string = row["date_publication"]
    
    # Vérifier que la string peut être correctement traitée
    if "il y a" in publication_date_string :

        days_offset = int(publication_date_string.split(" ")[-2])
        effective_date = datetime.datetime.now() - datetime.timedelta(days=days_offset)
        formatted_date = effective_date.strftime("%d/%m/%Y")
        data.at[i, "date_publication"] = formatted_date
        


# Enregistrer le DataFrame avec la colonne "date effective" remplie
#data.to_csv("data_with_effective_date.csv", index=False)
```
